# User Based

```python
test_df_1 = test_df.groupby(['item_id']).size().reset_index().rename(columns={0:'cnt'}).sort_values(by=['cnt'],ascending=False)
test_df['item_id'] = test_df['item_id'].astype('str')
test_df_1['item_id'] = test_df_1['item_id'].astype('str')
test_df_2 = test_df[['buyer_admin_id','item_id']].groupby(['buyer_admin_id']).agg(lambda x: ' '.join(x.values))
print(test_df_2.shape[0])
test_df_2.head()
```

```python
test_df_2
```

```python
train_yy_1 = train_yy.groupby(['item_id']).size().reset_index().rename(columns={0:'cnt'}).sort_values(by=['cnt'],ascending=False)
train_yy['item_id'] = train_yy['item_id'].astype('str')
train_yy_1['item_id'] = train_yy_1['item_id'].astype('str')
train_yy_1_1 = train_yy_1[train_yy_1['cnt']>=3]
train_yy = train_yy[train_yy['item_id'].isin(train_yy_1_1['item_id'])]
train_yy_2 = train_yy[['buyer_admin_id','item_id']].groupby(['buyer_admin_id']).agg(lambda x: ' '.join(x.values))
print(train_yy_2.shape[0])
train_yy_2.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vec = CountVectorizer()
train_yy_v = vec.fit_transform(train_yy_2['item_id'])
train_yy_v
```

```python
test_df_v = vec.transform(test_df_2['item_id'])
test_df_v
```

```python

import numpy as np

result = []
for i in range(test_df_v.shape[0]):
    test_i = test_df_v[i]
    test_1 = train_yy_v * test_i.T
    test_1 = test_1.toarray()
    corr_mat = np.argsort(-test_1.T,axis=1)
#     print(corr_mat)
#     corr_top_lst = corr_mat[:,0:2][0]
#     print(train_xx_1.iloc[corr_top_lst,:].values)
#     print(" ".join(train_xx_1.iloc[corr_top_lst,:].values.flatten()))
    j = 1
    top_n = 0
    top_30 = 30
    item_top = []
    while top_n<top_30:
        corr_top_lst = corr_mat[:,0:j][0]
        item_lst_clue = " ".join(train_yy_2.iloc[corr_top_lst,:].values.flatten()).split(" ")
        item_lst_have = test_df.iloc[i,].values[0].split(" ")
        item_top = set(item_lst_clue) - set(item_lst_have)
        top_n = len(item_top)
        j = j+1
#         print(item_lst_clue)
#         print(item_lst_have)
#         print(top_n)
    result.append([test_df_2.iloc[i,].name] + list(item_top)[0:30])

result = pd.DataFrame(result)
result
```

# 组合候选集

```python
result_con = []
for idx,item in test_df_2.iterrows():
    tmp_lst = tmp2[tmp2['buyer_admin_id']==item.name]['item_id'].values.tolist()
    tmp_lst = tmp_lst + hot_item_10.index.values.tolist()
    tmp_lst = tmp_lst + result[result[0]==item.name].iloc[:,1:].values.tolist()[0]
    result_con.append([item.name]+tmp_lst)
result_con = pd.DataFrame(result_con)
```
