# 解析图片的标注

## 限制品包括：铁壳打火机、黑钉打火机、刀具、电池电容以及剪刀五类（类别id依次从1到5）。
## 数据集标注采用json文件提供，json文件中的信息与coco数据集兼容，详见http://cocodataset.org/#format-data 中的object deteciton部分。其中，area和segmentation字段初赛为空，将在复赛提供。iscrowd字段为0，无意义。
## 需要注意的是：Bbox字段：[x,y,width,height]，为标准常用bounding box坐标 。新增的minAreaRect字段表示目标轮廓的最小外接矩形，依次为其四个顶点坐标。如果图片中没有违禁品，则显示"rects": []。

```python
import json

label_data = {}
with open('../data/tt1/jinnan2_round1_train_20190305/train_no_poly.json', 'r') as load_1:
    label_data = json.load(load_1)
```

## 读一张图片，加上标注看看 (一张图片可能有多个标注)

```python
import cv2

img_name = label_data['images'][4]['file_name']
print(img_name)
img1 = cv2.imread('../data/tt1/jinnan2_round1_train_20190305/restricted/'+img_name)
cv2.namedWindow('img1', cv2.WINDOW_NORMAL)
cv2.imshow('img1', img1)
cv2.waitKey(0)
```

```python
# 用pandas
```

```python
import pandas as pd
images = []
for i in range(len(label_data['images'])):
    images.append([label_data['images'][i]['id'],label_data['images'][i]['file_name']])
images = pd.DataFrame(images,columns=['id','file_name'])
images.head()
```

```python
annotations = []
for i in range(len(label_data['annotations'])):
    ann_obj = label_data['annotations'][i]
    annotations.append((ann_obj['area'],ann_obj['bbox'],ann_obj['category_id'],ann_obj['id'],ann_obj['image_id'],ann_obj['iscrowd'],ann_obj['minAreaRect'],ann_obj['segmentation']))
annotations = pd.DataFrame(annotations,columns=['area','bbox','category_id','id','image_id','iscrowd','minAreaRect','segmentation'])
annotations.head()
```

```python
img_id = images.iloc[4]['id']
print(img_id)
img_name = images.iloc[4]['file_name']
print(img_name)
ann_img_ids = annotations[annotations['image_id']==img_id]
print(ann_img_ids)
```

```python
img1 = cv2.imread('../data/tt1/jinnan2_round1_train_20190305/restricted/'+img_name)
cv2.namedWindow('img1', cv2.WINDOW_NORMAL)
for i in range(len(ann_img_ids)):
    bbox = ann_img_ids.iloc[i]['bbox']
    cv2.rectangle(img1,(bbox[0],bbox[1]),(bbox[0]+bbox[2],bbox[1]+bbox[3]),(0,255,0),2) #矩形，指定左上和右下点的坐标，颜色，粗细
cv2.imshow('img1', img1)
cv2.waitKey(0)
```
