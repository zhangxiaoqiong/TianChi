# 特征数据

```python
feature_data = train_data[['item_price_level', 'item_sales_level',
       'item_collected_level', 'item_pv_level', 'shop_review_num_level', 'shop_review_positive_rate',
       'shop_star_level', 'shop_score_service', 'shop_score_delivery',
       'shop_score_description','is_trade']]

feature_data = feature_data[(feature_data['item_sales_level']>-1) & (feature_data['shop_review_positive_rate']>-1) & (feature_data['shop_score_service']>-1) & (feature_data['shop_score_delivery']>-1) & (feature_data['shop_score_description']>-1)]
feature_data = feature_data.drop_duplicates()
```

```python
feature_data.head()
```

```python
feature_data['is_trade'].sum()
```

# 训练预测模型

```python
from sklearn.cross_validation import train_test_split

X_train, X_test,y_train,y_test = train_test_split(feature_data[['item_price_level', 'item_sales_level',
       'item_collected_level', 'item_pv_level', 'shop_review_num_level', 'shop_review_positive_rate',
       'shop_star_level', 'shop_score_service', 'shop_score_delivery',
       'shop_score_description']],feature_data[['is_trade']], test_size = 0.2)

```

```python
X_train.head()
```

```python
y_train.head()
```

```python
from sklearn.linear_model import LogisticRegression

LRmodel = LogisticRegression()
LRmodel.fit_transform(X=X_train,y=y_train)
```

```python
predict_y = LRmodel.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report 
# from sklearn.metrics import precision_recall_curve, roc_curve, auc 

# precision, recall, thresholds = precision_recall_curve(y_true=y_test, predict_y)

print(confusion_matrix(y_true=y_test,y_pred=predict_y))
```

```python
print(classification_report(y_pred=predict_y,y_true=y_test))
```

```python
#评估
import scipy as sp

def logloss(act, pred):
    epsilon = 1e-15
    pred = sp.maximum(epsilon, pred)
    pred = sp.minimum(1-epsilon, pred)
    ll = sum(act*sp.log(pred) + sp.subtract(1,act)*sp.log(sp.subtract(1,pred)))
    ll = ll * -1.0/len(act)
    return ll
```

```python
predict_proba_y = LRmodel.predict_proba(X_test)[:,1]
```

```python
#持久化模型
from sklearn.externals import joblib

joblib.dump(LRmodel,'./LR.model')
```

# 预测

```python
test_feature_data = testa_data[['item_price_level', 'item_sales_level',
       'item_collected_level', 'item_pv_level', 'shop_review_num_level', 'shop_review_positive_rate',
       'shop_star_level', 'shop_score_service', 'shop_score_delivery',
       'shop_score_description']]

test_predict_proba_y = LRmodel.predict_proba(test_feature_data)
```

```python
predicted_score_y = pd.DataFrame(test_predict_proba_y[:,1],columns=['predicted_score'])
```

```python
result = pd.concat([testa_data[['instance_id']],predicted_score_y],axis=1)
```
