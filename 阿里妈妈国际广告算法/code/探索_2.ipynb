```python
feature_data = train_data[['item_price_level', 'item_sales_level','item_collected_level', 'item_pv_level', 
                           'shop_review_num_level', 'shop_review_positive_rate','shop_star_level', 'shop_score_service', 'shop_score_delivery','shop_score_description',
                           'user_gender_id', 'user_age_level', 'user_occupation_id', 'user_star_level',
                           'context_page_id','is_trade']]

feature_data = feature_data[feature_data['item_price_level']>-1]
feature_data = feature_data[feature_data['item_sales_level']>-1]
feature_data = feature_data[feature_data['item_collected_level']>-1]
feature_data = feature_data[feature_data['item_pv_level']>-1]
feature_data = feature_data[feature_data['shop_review_num_level']>-1]
feature_data = feature_data[feature_data['shop_review_positive_rate']>-1]
feature_data = feature_data[feature_data['shop_star_level']>-1]
feature_data = feature_data[feature_data['shop_score_service']>-1]
feature_data = feature_data[feature_data['shop_score_delivery']>-1]
feature_data = feature_data[feature_data['shop_score_description']>-1]
feature_data = feature_data[feature_data['user_gender_id']>-1]
feature_data = feature_data[feature_data['user_age_level']>-1]
feature_data = feature_data[feature_data['user_occupation_id']>-1]
feature_data = feature_data[feature_data['user_star_level']>-1]
feature_data = feature_data[feature_data['context_page_id']>-1]
feature_data = feature_data.drop_duplicates()
feature_data.head()
```

```python
feature_data['shop_star_level'] = feature_data['shop_star_level']-5000
feature_data['user_age_level'] = feature_data['user_age_level']-1000
feature_data['user_occupation_id'] = feature_data['user_occupation_id']-2000
feature_data['user_star_level'] = feature_data['user_star_level']-3000
feature_data['context_page_id'] = feature_data['context_page_id']-4000
```

```python
feature_data.head()
```

```python
from sklearn.cross_validation import train_test_split

X_train, X_test,y_train,y_test = train_test_split(feature_data.drop(['is_trade'],axis=1),feature_data[['is_trade']], test_size = 0.2)
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegression

LRmodel = LogisticRegression()
LRmodel.fit_transform(X=X_train,y=y_train)
```

```python
predict_y = LRmodel.predict(X_test)

from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report 
# from sklearn.metrics import precision_recall_curve, roc_curve, auc 

# precision, recall, thresholds = precision_recall_curve(y_true=y_test, predict_y)

print(confusion_matrix(y_true=y_test,y_pred=predict_y))
```

```python
predict_proba_y = LRmodel.predict_proba(X_test)[:,1]
```

```python
predict_proba_y
```

```python
#评估
import scipy as sp

def logloss(act, pred):
    epsilon = 1e-15
    pred = sp.maximum(epsilon, pred)
    pred = sp.minimum(1-epsilon, pred)
    ll = sum(act*sp.log(pred) + sp.subtract(1,act)*sp.log(sp.subtract(1,pred)))
    ll = ll * -1.0/len(act)
    return ll
```

```python
from sklearn.metrics import log_loss

log_loss(y_true=y_test,y_pred=predict_proba_y)
```
